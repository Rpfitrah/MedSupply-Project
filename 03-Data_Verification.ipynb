Check Column Name Length and Escaped Characters

```python
%%sql
SELECT column_name,
       length(column_name) AS len,
       encode(column_name::bytea, 'escape') AS escaped
FROM information_schema.columns
WHERE table_name = 'customers'
ORDER BY ordinal_position;
```

```python
%%sql

SELECT 
    mp.product_code,
    mp.product_name,
    mp.inden_lead_time_days,
    mp.stock_available,
    ti.inden_actual_time_days
FROM master_products as mp
JOIN transactions_item as ti USING (product_code)
WHERE ti.inden_flag = 'Yes'
LIMIT 5;
```

### Missing value Percentage

```python
%%sql

SELECT 
    (COUNT(*) - COUNT(supplier_cost)) * 100.0 / COUNT(*) AS missing_percentage
FROM master_products;
```

```python
%%sql

SELECT 
    (COUNT(*) - COUNT(inden_lead_time_days)) * 100.0 / COUNT(*) AS missing_percentage
FROM master_products;
```

```python
%%sql

SELECT 
    (COUNT(*) - COUNT(stock_available)) * 100.0 / COUNT(*) AS missing_percentage
FROM master_products;
```

```python
%%sql

SELECT 
    (COUNT(*) - COUNT(unit_price)) * 100.0 / COUNT(*) AS missing_percentage
FROM transactions_item;
```

Status unik dari invoice_status

```python
%%sql

SELECT 
    DISTINCT invoice_status as unique_status,
    COUNT(*) AS unique_count
FROM transactions_main
GROUP BY invoice_status;
```
